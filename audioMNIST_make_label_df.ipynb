```python
from glob import glob
import pandas as pd
```

```python
file_paths = glob('data/AudioMNIST/audio/**/*.wav', recursive=True)
```

```python
file_paths
```

```python
file_paths[0].split('/')[-1].split('_')[0]
```

```python
labels = []
relative_paths = []
persons = []

for file in file_paths:
    name = file.split('/')[-1]
    person = file.split('/')[-2]
    label = name.split('_')[0]
    labels.append(label)
    relative_paths.append(file)
    persons.append(person)
```

```python
df = pd.DataFrame()
```

```python
df['label'] = labels
```

```python
df['person'] = persons
```

```python
df["relative_path"] = relative_paths
```

```python
df.head()
```

```python
df.to_csv('data/AudioMNIST/audioMNIST_meta.csv')
```
